## Etape 1

```java
import java.util.Random;

// Étape 1 - Tirage aléatoire
Random hasard = new Random();
int alea = hasard.nextInt(6) + 1;

System.out.println("Valeur du dé : " + alea);
System.out.println();

// Étape 2 - Affichage du dé
char u, d, q, s;

// u vaut '*' quand la valeur du dé est impaire
if (alea % 2 == 1) {
    u = '*';
} else {
    u = ' ';
}

// d vaut '*' quand la valeur du dé vaut ou dépasse 2
if (alea >= 2) {
    d = '*';
} else {
    d = ' ';
}

// q vaut '*' quand la valeur du dé vaut ou dépasse 4
if (alea >= 4) {
    q = '*';
} else {
    q = ' ';
}

// s vaut '*' quand la valeur du dé vaut ou dépasse 6
if (alea >= 6) {
    s = '*';
} else {
    s = ' ';
}

// Affichage du dé avec contours
System.out.println("+-------+");
System.out.println("| " + q + "   " + d + " |");
System.out.println("| " + s + " " + u + " " + s + " |");
System.out.println("| " + d + "   " + q + " |");
System.out.println("+-------+");
```

# Bonus : ajout d'un dé à côté 

```java
import java.util.Random;

Random hasard = new Random();

// PREMIER DÉ 
int alea1 = hasard.nextInt(6) + 1;

char u1, d1, q1, s1;

if (alea1 % 2 == 1) {
    u1 = '*';
} else {
    u1 = ' ';
}

if (alea1 >= 2) {
    d1 = '*';
} else {
    d1 = ' ';
}

if (alea1 >= 4) {
    q1 = '*';
} else {
    q1 = ' ';
}

if (alea1 >= 6) {
    s1 = '*';
} else {
    s1 = ' ';
}

// DEUXIÈME DÉ
int alea2 = hasard.nextInt(6) + 1;

char u2, d2, q2, s2;

if (alea2 % 2 == 1) {
    u2 = '*';
} else {
    u2 = ' ';
}

if (alea2 >= 2) {
    d2 = '*';
} else {
    d2 = ' ';
}

if (alea2 >= 4) {
    q2 = '*';
} else {
    q2 = ' ';
}

if (alea2 >= 6) {
    s2 = '*';
} else {
    s2 = ' ';
}

// ===== AFFICHAGE DES DEUX DÉS CÔTE À CÔTE =====
System.out.println("Dé 1 : " + alea1 + "        Dé 2 : " + alea2);
System.out.println();

System.out.println("+-------+    +-------+");
System.out.println("| " + q1 + "   " + d1 + " |    | " + q2 + "   " + d2 + " |");
System.out.println("| " + s1 + " " + u1 + " " + s1 + " |    | " + s2 + " " + u2 + " " + s2 + " |");
System.out.println("| " + d1 + "   " + q1 + " |    | " + d2 + "   " + q2 + " |");
System.out.println("+-------+    +-------+");



```
